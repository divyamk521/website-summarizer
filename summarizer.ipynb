```python
import requests
from bs4 import BeautifulSoup
import ollama
from IPython.display import Markdown, display
```

```python
def get_website_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    for tag in soup(["script", "style", "nav", "footer"]):
        tag.decompose()
    text = soup.get_text(separator="\n", strip=True)
    return text[:2000]
```

```python

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]

print(user_prompt_prefix)
```

```python
def summarize(url):
    print(f"Fetching: {url}\n")
    website = get_website_text(url)
    response = ollama.chat(
        model="tinyllama",
        messages=messages_for(website)
    )
    return response["message"]["content"]  # ← return instead of print
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

url = input("Enter the website URL: ")

print("\nHow do you want the summary?")
print("1. 5 bullet points")
print("2. Short paragraph")
print("3. Key facts only")
style = input("\nChoose 1, 2 or 3: ")

styles = {
    "1": "Summarize this in 5 bullet points",
    "2": "Summarize this in one short paragraph",
    "3": "List only the key facts from this content"
}

chosen_prompt = styles.get(style, "Summarize this in 5 bullet points")
user_prompt_prefix = chosen_prompt + "\n\nImportant: Always respond in English only, regardless of the website language.\n\nHere are the contents of the website:\n"

display_summary(url)
```
